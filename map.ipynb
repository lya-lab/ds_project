```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature

#  PlateCarree 
fig = plt.figure(figsize=(8, 10))
ax = plt.axes(projection=ccrs.PlateCarree())

# 
ax.set_extent([-10, 2, 49, 61], crs=ccrs.PlateCarree())

# 
ax.coastlines(resolution='10m')
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')

# 
areas = {
    'NZ24': (-1.5926, 54.784),
    'NY30': (-2.9775, 54.42),
    'NY75': (-2.3557, 54.906)
}

for area, (lon, lat) in areas.items():
    ax.plot(lon, lat, marker='o', color='red', markersize=5, transform=ccrs.PlateCarree())
    ax.text(lon + 0.1, lat + 0.1, area, fontsize=12, transform=ccrs.PlateCarree())

# 
plt.title('Map of Marked Areas')

plt.savefig('map.png', bbox_inches='tight')  #  PNG 
# show
plt.show()
```
